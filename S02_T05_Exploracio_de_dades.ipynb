Mostra clarament que on hi manca força informació és en reportar els **minuts de retràs per cada causa**. De 1936758 entrades en fallen 689270 (35,6% del total).
Hi han alguns registres que li falta la **hora d'arribada real**, amb el que no es pot treballar amb les dades de retrasos. Elimino aquests registres (7110) ja que només suposen el 0,37% de les dades totals. 

```python
indexDel=list(df[df['ArrTime'].isnull()].index)#index de les línees que s'han de borrar
df.drop(indexDel, inplace=True)
```

```python
df.isnull().sum() #suma els valors nuls de cada columna i els reporta.
```

Eliminem també les que falten dades de temps de vol o retràs en l'arrivada que manquen i són molt poques respecte el total.

```python
indexDel=list(df[df['AirTime'].isnull()].index)#index de les línees que s'han de borrar
df.drop(indexDel, inplace=True)
df.isnull().sum()
```

Canviem el valor del número d'avió (TailNum) a 'Unkown' pels 3 registres que no el tenen definit.


```python
df['TailNum'].fillna(value='Unknown',inplace=True)
df.isnull().sum()
```

```python
df.groupby('Cancelled').count()
```

Al netejar el dataframe no queden vols cancelats. Es poden borrar totes les columnes referents a la cancelació i desviació de vols.

```python
df.drop(['Cancelled','CancellationCode', 'Diverted'],axis=1,inplace=True)
```

Finalment generarem un sub-dataframe més petit (2/3 del  total) que contingui les dades de les causes del retràs per quan es vulgui estudiar quines són les causes dels retards.

```python
df_cause=df.copy()
indexDel=list(df_cause[df_cause['NASDelay'].isnull()].index)#index de les línees que s'han de borrar
df_cause.drop(indexDel, inplace=True)
df_cause.isnull().sum()
```

```python
#Com ja tenim un dataframe per estudiar especificament els valors de retràs reportats, borrem les olumnes d'aquest dataframe.
df.drop(['CarrierDelay','WeatherDelay', 'NASDelay', 'SecurityDelay', 'LateAircraftDelay'],axis=1,inplace=True)
df.isnull().sum()
```

### Crear Columnes noves

```python
df['AverageSpeed']=df['Distance']/(df['AirTime']/60)#Dividim entre 60 perquè són minuts i volem per hora
```

```python
df['AverageSpeed'].head(20)
```

```python
df['Delayed']=df['ArrTime']>df['CRSArrTime'] #Generem columna de si hi ha retràs (=True) o no (False)
```

```python
df['UniqueCarrier'].value_counts().head(3)
```
